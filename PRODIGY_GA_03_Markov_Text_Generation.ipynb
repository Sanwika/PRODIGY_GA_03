```python
import random

text = """
Artificial Intelligence is transforming the world.
AI is used in healthcare, education, and transportation.
Generative AI can create text, images, and music.
"""

words = text.split()

markov_chain = {}

for i in range(len(words) - 1):
    word = words[i]
    next_word = words[i + 1]

    if word not in markov_chain:
        markov_chain[word] = []

    markov_chain[word].append(next_word)

current_word = random.choice(words)
generated_text = [current_word]

for _ in range(30):
    if current_word in markov_chain:
        current_word = random.choice(markov_chain[current_word])
        generated_text.append(current_word)
    else:
        break

print("Generated Text:")
print(" ".join(generated_text))
```
